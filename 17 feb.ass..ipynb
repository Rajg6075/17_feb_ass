```python
Q1. What is MongoDB? Explain non-relational databases in short. In which scenarios it is preferred to use
MongoDB over SQL databases?

ANS= MongoDB: MongoDB is a popular open-source NoSQL database management system that falls under the category of non-relational databases.
It is designed to store and manage unstructured or semi-structured data in a flexible and scalable manner.

Non-Relational Databases: Non-relational databases, also known as NoSQL databases, depart from the traditional relational model used in SQL databases. 
They provide more flexible data models to handle various types of data, such as documents, key-value pairs, columnar data, and graph structures. NoSQL databases prioritize scalability, high performance, and schema flexibility, making them suitable for scenarios where traditional relational databases might face limitations.

Preferred Scenarios for MongoDB (NoSQL) over SQL Databases:
Scalability and High Volume Data: MongoDB is well-suited for scenarios where there's a need to handle large volumes of rapidly changing and unstructured data. It can distribute data across multiple nodes easily, making it highly scalable.
Complex Data Structures: If your data doesn't fit neatly into the tabular structure of SQL databases, MongoDB's document model allows you to store hierarchical and complex data structures in a single document.
Agile Development: In rapidly changing environments, where requirements evolve quickly, MongoDB's schema-less or flexible schema model can be advantageous. Developers can adapt the data structure without changing the database schema.

Q2. State and Explain the features of MongoDB.

ANS= MongoDB is a widely-used NoSQL database management system that offers a range of features that make it suitable for various use cases. Here are some key features of MongoDB:

Document-Oriented:
MongoDB stores data in documents using a flexible, JSON-like format called BSON (Binary JSON). This allows complex and nested data structures to be stored as a single document, making it well-suited for semi-structured or unstructured data.
Schema Flexibility:
MongoDB's dynamic schema model enables you to store different fields in different documents without the need for a predefined schema. This flexibility is particularly useful in scenarios where data evolves or varies across documents.
High Scalability:
MongoDB can scale horizontally by distributing data across multiple servers or clusters. This sharding capability enables the database to handle large amounts of data and high traffic loads.

Q3. Write a code to connect MongoDB to Python. Also, create a database and a collection in MongoDB.

ANS= import pymongo

# Connect to MongoDB
client = pymongo.MongoClient("mongodb://localhost:27017/")

# Create a database
mydb = client["mydatabase"]

# Create a collection within the database
mycollection = mydb["mycollection"]

# Sample data to insert
data = {"name": "John", "age": 30, "city": "New York"}

# Insert data into the collection
insert_result = mycollection.insert_one(data)

# Print the inserted document's ID
print("Inserted document ID:", insert_result.inserted_id)

# Close the connection
client.close()

Q4. Using the database and the collection created in question number 3, write a code to insert one record,
and insert many records. Use the find() and find_one() methods to print the inserted record.

ANS= import pymongo

# Connect to MongoDB
client = pymongo.MongoClient("mongodb://localhost:27017/")
mydb = client["mydatabase"]
mycollection = mydb["mycollection"]

# Insert one record
one_record = {"name": "Alice", "age": 25, "city": "London"}
insert_result = mycollection.insert_one(one_record)
print("Inserted document ID:", insert_result.inserted_id)

# Insert many records
many_records = [
    {"name": "Bob", "age": 28, "city": "Paris"},
    {"name": "Eve", "age": 22, "city": "Berlin"}
]
insert_many_result = mycollection.insert_many(many_records)
print("Inserted IDs for many documents:", insert_many_result.inserted_ids)

# Find and print inserted record using find_one()
inserted_record = mycollection.find_one({"name": "Alice"})
print("Inserted record (find_one()):", inserted_record)

# Find and print all inserted records using find()
all_inserted_records = mycollection.find()
print("All inserted records (find()):")
for record in all_inserted_records:
    print(record)

# Close the connection
client.close()


Q5. Explain how you can use the find() method to query the MongoDB database. Write a simple code to
demonstrate this.

ANS= The find() method in MongoDB is used to query the database and retrieve documents that match specific criteria. It returns a cursor, which is an iterable object that can be used to access the matching documents. You can apply various query conditions, sorting, and projection to control what data you retrieve.
import pymongo

# Connect to MongoDB
client = pymongo.MongoClient("mongodb://localhost:27017/")
mydb = client["mydatabase"]
mycollection = mydb["mycollection"]

# Insert sample data
sample_data = [
    {"name": "Alice", "age": 25, "city": "London"},
    {"name": "Bob", "age": 28, "city": "Paris"},
    {"name": "Eve", "age": 22, "city": "Berlin"},
    {"name": "John", "age": 30, "city": "New York"}
]
mycollection.insert_many(sample_data)

# Query and print documents with age greater than 25
query = {"age": {"$gt": 25}}  # $gt stands for greater than
cursor = mycollection.find(query)

print("Documents with age greater than 25:")
for document in cursor:
    print(document)

# Close the connection
client.close()

Q6. Explain the sort() method. Give an example to demonstrate sorting in MongoDB.

ANS= The sort() method in MongoDB is used to specify the sorting order for the results of a query. It allows you to retrieve documents from a collection in a particular order based on one or more fields.
 sort(field, direction)

    import pymongo

# Connect to MongoDB
client = pymongo.MongoClient("mongodb://localhost:27017/")
mydb = client["mydatabase"]
mycollection = mydb["students"]

# Insert sample data
sample_data = [
    {"name": "Alice", "age": 20, "score": 85},
    {"name": "Bob", "age": 22, "score": 75},
    {"name": "Eve", "age": 21, "score": 90},
    {"name": "John", "age": 23, "score": 80}
]
mycollection.insert_many(sample_data)

# Query and sort documents by score in descending order
query = {}  # Empty query retrieves all documents
sort_field = "score"
sort_direction = -1
cursor = mycollection.find(query).sort(sort_field, sort_direction)

print("Documents sorted by score:")
for document in cursor:
    print(document)

# Close the connection
client.close()

Q7. Explain why delete_one(), delete_many(), and drop() is used.

ANS= delete_one() Method:
The delete_one() method is used to delete a single document from a collection that matches a specified query. If multiple documents match the query, only the first one encountered will be deleted.
result = collection.delete_one(query)

delete_many() Method:
The delete_many() method is used to delete multiple documents from a collection that match a given query. It removes all documents that satisfy the specified criteria.
result = collection.delete_many(query)

drop() Method:
The drop() method is used to completely remove an entire collection from the database. This operation is permanent and cannot be undone. It is useful when you want to delete all data from a collection and do not plan to use the collection anymore.
collection.drop()
```
